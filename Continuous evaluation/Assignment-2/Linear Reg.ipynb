```python
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
# convert boolean data to numerical data(gender)
dataMapping={"Female":1, "Male":2}
data['Gender'] = data['gender'].map(dataMapping)
data
```

```python
# convert Object data to numerical data(course)
#pd.to_numeric(data['course'])
#data
```

```python
# convert boolean data to numerical data(Marital Status)
dataMapping={"Yes":1, "No":2}
data['Marital status Modify'] = data['Marital status'].map(dataMapping)
data
```

```python
# convert boolean data to numerical data(Depression)
dataMapping={"Yes":1, "No":2}
data['Depression'] = data['Do you have Depression?'].map(dataMapping)
data
```

```python
# convert boolean data to numerical data(Do you have Anxiety?)
dataMapping={"Yes":1, "No":2}
data['Anxiety'] = data['Do you have Anxiety?'].map(dataMapping)
data
```

```python
# convert boolean data to numerical data(Panic attack)
dataMapping={"Yes":1, "No":2}
data['Panic attack'] = data['Do you have Panic attack?'].map(dataMapping)
data
```

```python
# convert boolean data to numerical data(Treatment)
dataMapping={"Yes":1, "No":2}
data['Treatment'] = data['Did you seek any specialist for a treatment?'].map(dataMapping)
data
```

```python
x=data[['Age']]
x
```

```python
y=data['Panic attack']
y
```

```python
model = LinearRegression()
```

```python
model.fit(x, y)
```

```python
model.predict([[18.0]])
```

```python
model.intercept_
```

```python
model.coef_
```

```python
y1=model.coef_*18.0+model.intercept_
y1
```

```python
plt.xlabel("Age")
plt.ylabel("Panic attack")
plt.scatter(x, y)
plt.plot(data.Age, model.predict(data[['Age']]), color='red')
```

```python
#R2 Method(for Singal variable)
model = LinearRegression()
X,y=x,y
model.fit(X, y)
model.score(X, y)*100
```
