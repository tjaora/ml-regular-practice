```python
import scipy.stats as stats
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
dataset = pd.read_csv("/content/Student Mental health1.csv")
dataset.head(13)
```

```python
dataset_table1=pd.crosstab(dataset['Do you have Depression?'],dataset['Do you have Panic attack?'])
print(dataset_table1)
```

```python
dataset_table2=pd.crosstab(dataset['Choose your gender'],dataset['Marital status'])
print(dataset_table2)
```

```python
dataset_table3=pd.crosstab(dataset['Do you have Panic attack?'],dataset['Did you seek any specialist for a treatment?'])
print(dataset_table3)
```

```python
#Observed Values 1
Observed_Values1 = dataset_table1.values 
print("Observed Values :-\n",Observed_Values1)
```

```python
val=stats.chi2_contingency(dataset_table1)
val
```

```python
Expected_Values=val[3]
```

```python
no_of_rows=len(dataset_table1.iloc[0:2,0])
no_of_columns=len(dataset_table1.iloc[0,0:2])
ddof=(no_of_rows-1)*(no_of_columns-1)
print("Degree of Freedom:-",ddof)
alpha = 0.05
```

```python
from scipy.stats import chi2
chi_square=sum([(o-e)**2./e for o,e in zip(Observed_Values1,Expected_Values)])
chi_square_statistic=chi_square[0]+chi_square[1]
     
```

```python
print("chi-square statistic:-",chi_square_statistic)
```

```python
critical_value=chi2.ppf(q=1-alpha,df=ddof)
print('critical_value:',critical_value)
```

```python
#p-value
p_value=1-chi2.cdf(x=chi_square_statistic,df=ddof)
print('p-value:',p_value)
print('Significance level: ',alpha)
print('Degree of Freedom: ',ddof)
print('p-value:',p_value)
```

```python
#if chi_square_statistic>=critical_value:
 #   print("Reject H0,There is a relationship between 2 categorical variables")
#else:
 #   print("Retain H0,There is no relationship between 2 categorical variables")
    
if p_value<=alpha:
    print("Reject H0,There is a relationship between 2 categorical variables")
else:
    print("Retain H0,There is no relationship between 2 categorical variables")
```
